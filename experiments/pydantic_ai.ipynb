```python
import logfire

logfire.configure()
logfire.info('Hello, {place}!', place='World')
```

```python
query = "What sky is blue , give me funny answers in 10 words   ?"
logfire.info("Agent query: {query}", query=query)
```

```python
pirate_agent = Agent(
    name="Pirate Agent",
    description="A pirate who loves to talk like a pirate and tell jokes about the sea.",
    model="groq:llama-3.1-8b-instant",
    system_prompt="You are a pirate who loves to talk like a pirate and tell jokes about the sea."
)

response = await pirate_agent.run(query)
print("hi")
print("the agent response is:", response)
logfire.info("Agent response: {response}", response=response)

# --- Token observability ---
if hasattr(response, 'usage'):
    logfire.info("Token usage: {usage}", usage=response.usage)
    print("Token usage:", response.usage)
else:
    print("Token usage info not available in response object.")
```
